```python
import requests
import json
```

```python
with open('arenas.json', 'r') as f:
    data = json.load(f)
data
```

```python
addresses = [f"{x['address']},{x['city']},{x['province']} {x['postal']},{x['country']}" for x in data['arenas']]
addresses
```

```python
# test on one address
address = addresses[0].replace(' ', '%20').replace(',','%2C')

API_KEY = "151409d7df284a77bb597204fbd2b974"

url = f"https://api.geoapify.com/v1/geocode/search?text={address}&apiKey={API_KEY}"

headers = {}
headers["Accept"] = "application/json"

r = requests.get(url, headers=headers).json()

lat = r['features'][0]['properties']['lat']
lon = r['features'][0]['properties']['lon']
lat,lon
```

```python
for i in range(len(data['arenas'])):
    arena = data['arenas'][i]
    address = f"{arena['address']},{arena['city']},{arena['province']} {arena['postal']},{arena['country']}"
    url = f"https://api.geoapify.com/v1/geocode/search?text={address}&apiKey={API_KEY}"
    headers = {}
    headers["Accept"] = "application/json"

    r = requests.get(url, headers=headers).json()

    lat = r['features'][0]['properties']['lat']
    lon = r['features'][0]['properties']['lon']
    data['arenas'][i]['lat'] = str(lat)
    data['arenas'][i]['lon'] = str(lon)
```

```python
for i in range(len(data['arenas'])):
    data['arenas'][i]['lat'] = str(data['arenas'][i]['lat'])
    data['arenas'][i]['lon'] = str(data['arenas'][i]['lon'])
with open('arenas.json', 'w') as f:
    json.dump({'arenas': data}, f)
```
